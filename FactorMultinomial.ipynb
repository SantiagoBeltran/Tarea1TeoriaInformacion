```python
def fact(n,m):
    if n > m:
        return n * fact(n - 1,m)
    else:
        return 1
```

```python
def factor(n0, N, m):
    y=np.zeros(len(n0))
    nMas=0.5*(N+m-n0)
    nMenos=0.5*(N-m-n0)
    g=np.zeros([len(n0),3])
    for i in range(len(n0)):
        m=np.array([nMenos[i],n0[i],nMas[i]])
        g[i]=sort(m)
        y[i]=fact(N,g[i][-1])/(fact(g[i][0],1)*fact(g[i][1],1))
    return y    
    
```

```python
plt.figure()
s=np.arange(0,1,0.001)
plt.plot(s,(1-s**2)/2, label=r"$f_0$")
plt.plot(s,(1+s)**2/4, label=r"$f_+$")
plt.plot(s,(1-s)**2/4, label=r"$f_-$")
plt.legend()
```

```python
def hDelta(N, e):
    x=np.arange(1,2**N+1,e)
    y=np.zeros(len(x))
    for i in range(len(x)):
        y[i]=np.log(int(2**N+1.9999-x[i]))/(N*np.log(2))
        #print(int(2**N+1.9999-x[i]))
    d=x/(2**N)
    #print(y)
    return d,y

def h100(N):
    x=np.arange(0,1,0.01)
    y=np.zeros(len(x))
    for i in range(len(x)):
        m=round(1+((1-x[i])*2**N),100000)
        y[i]=np.log2(m)/N

    return x,y
```

```python
plt.figure(figsize=(12,12))
plt.subplot(2,2,1)
plt.plot(hDelta(1, 0.001)[0],hDelta(1, 0.001)[1], label="N=1")
plt.xlabel(r"$\delta$")
plt.ylabel(r"$\frac{1}{N}H_\delta(X^N)$")
plt.legend()
plt.subplot(2,2,2)
plt.plot(hDelta(2, 0.001)[0],hDelta(2, 0.001)[1],label="N=2")
plt.xlabel(r"$\delta$")
plt.ylabel(r"$\frac{1}{N}H_\delta(X^N)$")
plt.legend()
plt.subplot(2,2,3)
plt.plot(hDelta(3, 0.001)[0],hDelta(3, 0.001)[1],label="N=3")
plt.xlabel(r"$\delta$")
plt.ylabel(r"$\frac{1}{N}H_\delta(X^N)$")
plt.legend()
plt.subplot(2,2,4)
plt.scatter(h100(100)[0],h100(100)[1], s=2, label="N=100")
plt.scatter(h100(90)[0],h100(90)[1], s=2, label="N=90")
plt.scatter(h100(200)[0],h100(200)[1], s=2, label="N=200")
plt.xlabel(r"$\delta$")
plt.ylabel(r"$\frac{1}{N}H_\delta(X^N)$")
plt.legend()
```
